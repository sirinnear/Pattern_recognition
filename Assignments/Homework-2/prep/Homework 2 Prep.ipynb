```python
%config InlineBackend.figure_format = 'retina'
import numpy as np
from math import pi, sin, cos
from matplotlib import pyplot as plt
import pandas as pd
```

# Problem 1

```python
np.random.seed(1234)
points = np.random.rand(1000,2)*2
def logistic(s):
    return 1/(1+np.exp(-s))

def p(point):
    th = pi/3
    cx, cy = 1, 0.8
    x, y = point
    a, b = 0.6, 0.25
    s = ((x-cx)*sin(th) - (y-cy)*cos(th))**2/a**2 + ((x-cx)*cos(th) + (y-cy)*sin(th))**2/b**2 - 1
    return logistic(1.5*s)

def clazz(point):
    return 1 if np.random.rand() > p(point) else -1
```

```python
xclass = np.array([clazz(p) for p in points])
```

```python
pmask = xclass > 0
mmask = xclass < 0
xs = points[:, 0]
ys = points[:, 1]
plt.plot(xs[pmask], ys[pmask], '.r')
plt.plot(xs[mmask], ys[mmask], '.b')
```

```python
data = np.column_stack([points, xclass])
df = pd.DataFrame(data, columns=['x1', 'x2', 'xclass'])
df.to_csv('../data/problem1.csv', index=False)
```

```python
np.random.seed(4321)
points = np.random.rand(1000,2)
pos = np.random.rand(20, 2)/5
plt.scatter(points[:, 0], points[:, 1], marker='.')
plt.scatter(pos[:, 0], pos[:, 1], marker='.')

points = np.vstack([points, pos])
clazz = np.array([-1]*1000 + [1]*20)
```

```python
data = np.column_stack([points, clazz])
df = pd.DataFrame(data, columns=['x1', 'x2', 'xclass'])
df.to_csv('../data/problem3.csv', index=False)
```
